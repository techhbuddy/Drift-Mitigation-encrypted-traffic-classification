```python
full_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
from sklearn.utils import shuffle
full_df = shuffle(full_df, random_state=42)
print("Combined dataset shape:", full_df.shape)
print("Binary label distribution:\n", full_df['binary_label'].value_counts())
```

```python
clean_df = full_df.copy()
numeric_cols = clean_df.select_dtypes(include=['float64', 'int64']).columns
numeric_cols = [col for col in numeric_cols if col != 'binary_label']
for col in numeric_cols:
    Q1 = clean_df[col].quantile(0.25)
    Q3 = clean_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    clean_df[col] = clean_df[col].clip(lower=lower_bound, upper=upper_bound)
print("Shape after capping outliers:", clean_df.shape)
print("Binary label distribution after capping:")
print(clean_df['binary_label'].value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
X = full_df.drop(columns=['binary_label', 'class1'])
y = full_df['binary_label']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Feature scaling complete.")
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit_transform(X)
k = 15
selector = SelectKBest(score_func=chi2, k=k)
X_selected = selector.fit_transform(X_minmax, y)
selected_features = X.columns[selector.get_support()].tolist()
print(f"Selected top {k} features:\n{selected_features}")
```

```python
X = full_df[selected_features]
y = full_df['binary_label']
print("Feature matrix shape:", X.shape)
print("Target distribution:\n", y.value_counts())
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)
print("Balanced feature matrix shape:", X_balanced.shape)
print("Balanced label distribution:\n", pd.Series(y_balanced).value_counts())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced
)
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from collections import deque
from sklearn.metrics import accuracy_score
import numpy as np
window_size = 1000
X_window = deque(maxlen=window_size)
y_window = deque(maxlen=window_size)
predictions = []
for i in range(len(X_train)):
    x = X_train.iloc[i].values
    y = y_train.iloc[i]
    if len(X_window) == window_size:
        knn = KNeighborsClassifier(n_neighbors=3)
        knn.fit(list(X_window), list(y_window))
        y_pred = knn.predict([x])[0]
        predictions.append(y_pred)
    X_window.append(x)
    y_window.append(y)
y_true = y_train[window_size:]
print("KNN with Sliding Window Accuracy on training stream:", accuracy_score(y_true, predictions))
```

```python
from river import neighbors, metrics
from collections import deque
window_size = 200
X_window = deque()
y_window = deque()
knn_test_model = neighbors.KNNClassifier(n_neighbors=5)
accuracy = metrics.Accuracy()
for i, row in X_test.iterrows():
    x = row.to_dict()
    y = y_test.loc[i]
    y_pred = knn_test_model.predict_one(x)
    if y_pred is not None:
        accuracy.update(y, y_pred)
    X_window.append(x)
    y_window.append(y)
    if len(X_window) > window_size:
        X_window.popleft()
        y_window.popleft()
    knn_test_model = neighbors.KNNClassifier(n_neighbors=5)
    for xw, yw in zip(X_window, y_window):
        knn_test_model.learn_one(xw, yw)
print("KNN with Sliding Window Accuracy on test stream:", accuracy.get())
```

```python
from river import neighbors, metrics
from river.drift import ADWIN
from collections import deque
import pandas as pd
import numpy as np
X_test_drift = X_test.copy().values
for i in range(X_test_drift.shape[1]):
    X_test_drift[:, i] *= (1 + 0.05 * np.arange(len(X_test_drift)) / len(X_test_drift))
X_test_drift_df = pd.DataFrame(X_test_drift, columns=X_test.columns)
window_size = 200
X_window = deque()
y_window = deque()
knn_model = neighbors.KNNClassifier(n_neighbors=5)
accuracy = metrics.Accuracy()
drift_detector = ADWIN()
drift_detected = False
for i, row in X_test_drift_df.iterrows():
    x = row.to_dict()
    y_true = y_test.iloc[i]

    y_pred = knn_model.predict_one(x)
    if y_pred is not None:
        accuracy.update(y_true, y_pred)
        if drift_detector.update(int(y_pred != y_true)) and not drift_detected:
            print(f"\nDrift detected at index {i}")
            drift_detected = True
            X_window.clear()
            y_window.clear()
            knn_model = neighbors.KNNClassifier(n_neighbors=5)
    X_window.append(x)
    y_window.append(y_true)
    if len(X_window) > window_size:
        X_window.popleft()
        y_window.popleft()
    knn_model = neighbors.KNNClassifier(n_neighbors=5)
    for xw, yw in zip(X_window, y_window):
        knn_model.learn_one(xw, yw)
print("KNN with Drift Detection Accuracy on drifted test stream:", accuracy.get())
```

```python
from river import neighbors, metrics
from river.drift import ADWIN
from collections import deque
import pandas as pd
import numpy as np
window_size = 200
n_neighbors = 5
X_window = deque()
y_window = deque()
knn_model = neighbors.KNNClassifier(n_neighbors=n_neighbors)
accuracy_pre = metrics.Accuracy()
for i, row in X_test.iterrows():
    x = row.to_dict()
    y_true = y_test.loc[i]
    y_pred = knn_model.predict_one(x)
    if y_pred is not None:
        accuracy_pre.update(y_true, y_pred)
    X_window.append(x)
    y_window.append(y_true)
    if len(X_window) > window_size:
        X_window.popleft()
        y_window.popleft()
    knn_model = neighbors.KNNClassifier(n_neighbors=n_neighbors)
    for xw, yw in zip(X_window, y_window):
        knn_model.learn_one(xw, yw)
print(f"\nPre-drift Accuracy: {accuracy_pre.get():.2%}")
X_test_drift = X_test.copy().values
for i in range(X_test_drift.shape[1]):
    X_test_drift[:, i] *= (1 + 0.05 * np.arange(len(X_test_drift)) / len(X_test_drift))
X_test_drift_df = pd.DataFrame(X_test_drift, columns=X_test.columns)
X_window = deque()
y_window = deque()
knn_model = neighbors.KNNClassifier(n_neighbors=n_neighbors)
accuracy_post = metrics.Accuracy()
drift_detector = ADWIN()
drift_detected = False
for i, row in X_test_drift_df.iterrows():
    x = row.to_dict()
    y_true = y_test.iloc[i]
    y_pred = knn_model.predict_one(x)
    if y_pred is not None:
        accuracy_post.update(y_true, y_pred)
        if drift_detector.update(int(y_pred != y_true)) and not drift_detected:
            print(f"\nDrift detected at index {i}")
            drift_detected = True
            X_window.clear()
            y_window.clear()
            knn_model = neighbors.KNNClassifier(n_neighbors=n_neighbors)
    X_window.append(x)
    y_window.append(y_true)
    if len(X_window) > window_size:
        X_window.popleft()
        y_window.popleft()
    knn_model = neighbors.KNNClassifier(n_neighbors=n_neighbors)
    for xw, yw in zip(X_window, y_window):
        knn_model.learn_one(xw, yw)
print(f"\nPost-drift Accuracy: {accuracy_post.get():.2%}")
```
